```python
infos = {'title':[], 'writer':[], 'illustrator':[], 'genre':[], 'story':[], 'platform':[], 'thumb':[], 'url':[]}

# 작품 번호가 몇 번부터 몇 번까지 있는지 확인이 되지 않아 0~3999까지 다 돌림
for i in range(4000):
    print(i, end='\r')
    try:
        jsonurl = f'https://gateway-kw.kakao.com/decorator/v1/decorator/contents/{i}/profile'
        raw = requests.get(jsonurl, headers = {"User-Agent" : "Mozilla/5.0"})
        json_data = raw.json()
        story = json_data['data']['synopsis']
        title = json_data['data']['title']
        genre = json_data['data']['seoKeywords'][2][1:]
        
        writer = ''
        illustrator = ''
        for author in json_data['data']['authors']:
            if author['type'] == 'AUTHOR':
                if writer == '':
                    writer += author['name']
                else:
                    writer += ','+author['name']
            elif author['type'] == 'ILLUSTRATOR':
                if illustrator == '':
                    illustrator += author['name']
                else:
                    illustrator += ','+author['name']
        
        url = 'https://webtoon.kakao.com/content/' + json_data['data']['seoId'] + '/' + str(json_data['data']['id'])

        r = requests.get(url)
        soup = BeautifulSoup(r.text, 'html.parser')
        
        try:
            thumb = str(soup.findAll("video")[0]).split('poster="')[-1].split('"')[0]
        except:
            thumb = str(soup.findAll('picture')[1].source).split('"')[1]
        
        infos['title'].append(title)
        infos['writer'].append(writer)
        infos['illustrator'].append(illustrator)
        infos['genre'].append(genre)
        infos['story'].append(story)
        infos['platform'].append('카카오웹툰')
        infos['thumb'].append(thumb)
        infos['url'].append(url)
        
        
    except:
        continue
    
```
