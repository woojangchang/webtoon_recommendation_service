```python
from __future__ import unicode_literals # 이게 뭘까
import requests
import pymongo
import pandas as pd
from selenium import webdriver
from selenium.webdriver.chrome.options import Options
# from scrapy.http import TextResponse
import urllib.request
import urllib
import getpass
import time
import re
# from google.cloud import vision
# from google.cloud.vision import types
import os
import io
from os import walk
from bs4 import BeautifulSoup
from selenium.webdriver.common.keys import Keys
```

```python
if not os.path.exists('./thumb'):
    os.mkdir('./thumb')

URL = 'https://page.kakao.com/main?categoryUid=10&subCategoryUid=1002'

driver = webdriver.Chrome(executable_path='C:\chromedriver')
driver.implicitly_wait(time_to_wait=1)
driver.get(URL)
time.sleep(1)

#요일별 데이터 가져오기
for i in range(1,8): 

    driver.find_element_by_xpath('//*[@id="root"]/div[3]/div[2]/ul/li['+str(i)+']/div').click()
    time.sleep(1)

    for key_down in range(0,20):
        driver.find_element_by_tag_name('body').send_keys(Keys.PAGE_DOWN)
        time.sleep(0.5)

    html = driver.page_source
    soup = BeautifulSoup(html, 'html.parser')
    webtoon_area = soup.select('#root > div.jsx-3157985592.mainContents.mainContents_pc > div.css-1sna24c > div.css-1saqd06 > div')

    #제목 수집
    title = soup.find_all('div', {'class': 'css-9prbtm'}) # 제목 수집
    titles = []
    for t in range(len(title)):
        titles.append(re.sub('[\\/:*?"<>|]','',title[t].get_text()))

    # print(titles)
    titles = titles[5:]
    # print(titles)
    print('제목: ', len(titles))

    #이미지 수집
    data_list=soup.findAll('div',{'class':'css-1saqd06'})

    li_list = []
    for data in data_list:
        li_list.extend(data.findAll('li')) #해당 부분을 찾아 li_list와 병합

    thumb = []

    for li in li_list:
        img = li.find('img')
        img_src = img['data-src']
        thumb.append('https:' + img_src)
    thumb = thumb[5:]
    # print(thumb)
    print('썸네일: ', len(thumb))

    for k in range(len(thumb)):
#         if not os.path.exists('./imgs/{}'.format(titles[k])):
#             os.mkdir('./imgs/{}'.format(titles[k]))

        urllib.request.urlretrieve(thumb[k], './thumb/{}.jpg'.format(titles[k]))

    driver.execute_script("window.scrollTo(0, -10000)") 
                                   
driver.close()
print('finish')
```
