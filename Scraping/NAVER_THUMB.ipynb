```python
from bs4 import BeautifulSoup
from pprint import pprint
import requests, re, os
from urllib.request import urlretrieve
```

```python
#연재중인 웹툰 썸네일 가져오기
html = requests.get("https://comic.naver.com/webtoon/weekday")
soup = BeautifulSoup(html.text, 'html.parser')
html.close()
 
data_list=soup.findAll('div',{'class':'col_inner'})
# pprint(data1_list)
 
li_list = []
for data in data_list:
    li_list.extend(data.findAll('li'))
#pprint(li_list)
```

```python
li_list[0]
```

```python
li_list[0].find('a')['href']
```

```python
d = {'title':[], 'thumb':[], 'url':[]}
img_url_list = []
for li in li_list:
    img = li.find('img')
    title = img['title']
    img_src = img['src']
    url = 'comic.naver.com' + li.find('a')['href']

    d['title'].append(title)
    d['thumb'].append(img_src)
    d['url'].append(url)
print('finish')
```

```python
import pandas as pd

df = pd.DataFrame(d)
df
```

```python
df.to_csv('naverurl.csv', index=False)
```
