```python
import time
day = 1

infos = {'title':[], 'writer':[], 'illustrator':[], 'genre':[], 'story':[], 'platform':[], 'thumb':[], 'url':[]}

# 월~일요일 데이터
while day < 8:
    page = 0
    is_end = False
    while True:
        jsonurl = f"https://api2-page.kakao.com/api/v2/store/day_of_week_top/list?category=10&subcategory=0&page={page}&day={day}&bm=A"
        raw = requests.get(jsonurl, headers = {"User-Agent" : "Mozilla/5.0"})
        json_data = raw.json()
        is_end =json_data['is_end']
        
        lists = json_data['list']
        
        startnum = 0
        if page == 0:
            startnum = 5
        for i in range(startnum, len(lists)):
            title = json_data['list'][i]['title']
            genre = json_data['list'][i]['sub_category_title']
            story = json_data['list'][i]['description'].replace('&#34;', '"')
#             author = json_data['list'][i]['author']
            thumb = 'http://dn-img-page.kakao.com/download/resource?kid=' + json_data['list'][i]['image'] + '&filename=th2'
            url = 'https://page.kakao.com/home?seriesId=' + str(json_data['list'][i]['series_id'])
            
            
            driver.get(url)
            driver.find_element_by_css_selector('#root > div.jsx-885677927.mainContents.mainContents_pc.hiddenMenuContent > div > div > \
                                                div.css-sgpdfd > div > div.css-1ydjg2i > div.css-1nlm7ol > div.css-82j595 > button.css-zkp4tp').click()
            time.sleep(0.5)
            
            soup = BeautifulSoup(driver.page_source, 'html.parser')
            author_list = soup.find_all('span', 'jsx-3755015728')
            
            writer = ''
            illustrator = ''
            for i in range(len(author_list)//2):
                if author_list[2 * i].text == '글':
                    writer += author_list[2*i+1].text
                elif author_list[2 * i].text == '그림':
                    illustrator += author_list[2*i+1].text
                elif author_list[2 * i].text == '원작':
                    origin = author_list[2*i+1].text
                    if origin in writer:
                        continue
                    elif writer == '':
                        writer = origin
                    else:
                        writer += ',' + origin
            
            if writer == '':
                writer = illustrator
            if illustrator == '':
                illustrator = writer
            
            if len(author_list) == 1:
                writer = illustrator = author_list[0].text
            
            infos['title'].append(title)
#             infos['author'].append(author)
            infos['genre'].append(genre)
            infos['story'].append(story)
            infos['platform'].append('카카오페이지')
            infos['thumb'].append(thumb)
            infos['url'].append(url)
            infos['writer'].append(writer)
            infos['illustrator'].append(illustrator)
            
        print(day, page)
        # 요일의 마지막 페이지면 다음 요일로 넘어감
        if is_end:
            break
        page += 1
        
    day += 1
```
