```python
co = df[['score', 'score_num', 'best_recomm', 'best_unrecomm', 'comment_num', 'star']].corr()
co
```

```python
import seaborn as sns

plt.figure(figsize=(8, 6))
sns.heatmap(co, vmin=-1, cmap='RdBu_r', annot=True, fmt='.4f')
plt.show()
```
